```python

N = int(input("Enter the range limit (N): "))
D = int(input("Enter the divisor (D): "))

if N < 1:
    print("Range limit must be greater than zero.")
elif D == 0:
    print("Divisor cannot be zero.")
else:
    divisible_numbers = []

    for number in range(1, N + 1):
        if number % D == 0:
            divisible_numbers.append(number)

    print(f"Count of numbers divisible by {D}: {len(divisible_numbers)}")
    print("Divisible numbers:", divisible_numbers)
```
