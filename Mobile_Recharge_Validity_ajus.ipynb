```python
from datetime import datetime, timedelta

# Get dates from the user
recharge_date_input = input("Enter recharge date (YYYY-MM-DD): ")
today_date_input = input("Enter today's date (YYYY-MM-DD): ")

try:
    recharge_date = datetime.strptime(recharge_date_input, "%Y-%m-%d").date()
    today_date = datetime.strptime(today_date_input, "%Y-%m-%d").date()

    expiry_date = recharge_date + timedelta(days=28)

    if today_date < recharge_date:
        print("Invalid input: Today's date cannot be before the recharge date.")
    elif today_date < expiry_date:
        remaining_days = (expiry_date - today_date).days
        print(f"Plan is still valid. Remaining days: {remaining_days}")
    else:
        print("Plan expired. Recharge again.")

except ValueError:
    print("Invalid date format! Please use YYYY-MM-DD.")
```
