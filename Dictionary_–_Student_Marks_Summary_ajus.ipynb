```python

students = {}

# Get names and marks of three students
for i in range(3):
    name = input("Enter name: ")
    marks = float(input("Enter marks: "))
    students[name] = marks

# Calculate average marks
average_marks = sum(students.values()) / len(students)

print("Student Marks:", students)
print(f"Average Marks: {average_marks:.1f}")
```
