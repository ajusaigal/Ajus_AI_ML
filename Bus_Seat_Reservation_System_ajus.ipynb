```python

booked = ["Amit", "Sara", "John"]

cancel_name = input("Enter passenger name to cancel: ")
new_name = input("Enter new passenger name: ")
position = int(input("Enter booking position: "))

# Cancel the booking
if cancel_name in booked:
    booked.remove(cancel_name)
    print(f"{cancel_name}'s booking cancelled.")
else:
    print("Passenger name not found.")

# Insert the new passenger at the specified position
if 0 <= position <= len(booked):
    booked.insert(position, new_name)
else:
    print("Invalid position.")

print("Updated passenger list:", booked)
```
